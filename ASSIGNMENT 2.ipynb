```python
# Program (i): Even Numbers Between 1 and 50

for i in range(1, 51):
    if i % 2 != 0:      # skip odd numbers
        continue
    if i % 10 == 0:     # skip multiples of 10
        continue
    if i > 40:          # stop if number greater than 40
        break
    print(i)
```

```python
# Program (ii): Palindrome, Armstrong & Fibonacci Series

# Palindrome Check
num = 121
temp = num
rev = 0
while temp > 0:
    digit = temp % 10
    rev = rev * 10 + digit
    temp = temp // 10
if num == rev:
    print(num, "is a Palindrome")
else:
    print(num, "is not a Palindrome")

# Armstrong Check
num2 = 153
temp = num2
sum = 0
while temp > 0:
    digit = temp % 10
    sum = sum + digit**3
    temp = temp // 10
if num2 == sum:
    print(num2, "is an Armstrong number")
else:
    print(num2, "is not an Armstrong number")

# Fibonacci Series
n = 7
a, b = 0, 1
print("Fibonacci Series:")
for i in range(n):
    print(a)
    c = a + b
    a = b
    b = c
```

```python
# Program (iii): Factorial using Recursion (Module)

# File: mymodule.py
def factorial(n):
    if n == 0:
        return 1
    else:
        return n * factorial(n - 1)

# Main program
num = 5
print("Factorial of", num, "is", factorial(num))
```

# Output Example:
# Factorial of 5 is 120
