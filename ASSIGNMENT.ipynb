## (i) Perform all arithmetic operations on two numbers

````python
# Program to perform all arithmetic operations on two numbers

a = int(input("Enter first number: "))
b = int(input("Enter second number: "))

print("Addition:", a + b)
print("Subtraction:", a - b)
print("Multiplication:", a * b)
print("Division:", a / b)
print("Floor Division:", a // b)
print("Modulus:", a % b)
print("Exponent:", a ** b)
````

### ✅ Sample Output
```
Enter first number: 10
Enter second number: 5
Addition: 15
Subtraction: 5
Multiplication: 50
Division: 2.0
Floor Division: 2
Modulus: 0
Exponent: 100000
```

## (ii) Interchange list items (with & without temporary variable)

````python
# With temporary variable
list1 = [10, 20]
print("Before swapping (with temp):", list1)
temp = list1[0]
list1[0] = list1[1]
list1[1] = temp
print("After swapping (with temp):", list1)

# Without temporary variable
list2 = [30, 40]
print("\nBefore swapping (without temp):", list2)
list2[0], list2[1] = list2[1], list2[0]
print("After swapping (without temp):", list2)
````

### ✅ Sample Output
```
Before swapping (with temp): [10, 20]
After swapping (with temp): [20, 10]

Before swapping (without temp): [30, 40]
After swapping (without temp): [40, 30]
```

## (iii) Dictionary operations

````python
# Creating dictionary
student = {
    "name": "Rahul",
    "age": 20,
    "course": "Python"
}

print("Original Dictionary:", student)

# Add new element
student["marks"] = 85
print("After adding marks:", student)

# Remove element
student.pop("age")
print("After removing age:", student)

# Copy dictionary
copy_student = student.copy()
print("Copied Dictionary:", copy_student)

# Nested dictionary
student["details"] = {"city": "Delhi", "college": "ABC College"}
print("After adding nested dictionary:", student)
````

### ✅ Sample Output
```
Original Dictionary: {'name': 'Rahul', 'age': 20, 'course': 'Python'}
After adding marks: {'name': 'Rahul', 'age': 20, 'course': 'Python', 'marks': 85}
After removing age: {'name': 'Rahul', 'course': 'Python', 'marks': 85}
Copied Dictionary: {'name': 'Rahul', 'course': 'Python', 'marks': 85}
After adding nested dictionary: {'name': 'Rahul', 'course': 'Python', 'marks': 85, 'details': {'city': 'Delhi', 'college': 'ABC College'}}
```
